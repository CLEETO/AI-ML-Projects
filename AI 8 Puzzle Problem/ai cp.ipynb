```python
import random
import copy
import sys
depth=0
soln=[]
prob1=[]
f={}
graph={}
nodes=[]
import sys
def bfs():
    frontier=[]
    explored=[]
    global l
    l=[]
    global p
    p={}
    frontier.append(prob1)
    while frontier:
        node=frontier.pop(0)
        explored.append(node)
        if node==soln :
            print('Solution found')
            while node!=prob1:
                l.append(node)
                node=p[node]
            l.append(prob1)
            for i in reversed(l):
                display(i)
            sys.exit()
        for j in graph[node]:
            if j not in frontier and j not in explored:
                 frontier.append(j)
                 p[j]=node
def display(lis):
    for i in lis:
        print('-------------',end='\n|')
        for j in i:
            if len(j)==1:
              print(j,end='  |')
            else:
              print(j,end=' |')
        print()
    print('-------------')
def generate():
    global prob1
    print("Enter the values row wise(x for null):")
    for i in range(0,3):
        prob1.append(tuple([x if x!='x' else ' ' for x in input().split()]))
    graph[tuple(prob1)]=[]
    nodes.append(tuple(prob1))
    prob1=tuple(prob1)
def generatesoln():
    global soln
    print("Enter the values row wise(x for null):")
    for i in range(0,3):
        soln.append(tuple([x if x!='x' else ' ' for x in input().split()]))
    soln=tuple(soln)
def convertlist(prob):
    prob=list(prob)
    for i in range(3):
        prob[i]=list(prob[i])
    return prob
def converttup(prob):
    for i in range(3):
        prob[i]=tuple(prob[i])
    prob=tuple(prob)
    return prob
def checkh(prob):
    h=0
    for i in range(3):
        for j in range(3):
            if prob[i][j]!=soln[i][j] and prob[i][j]!=' ':
                h+=1
    return h        
def solve(prob):
    global depth
    depth+=1
    prob=convertlist(prob)
    ii=0
    ji=0
    for i in prob:
        if ' ' in i:
            ii=prob.index(i)
            ji=i.index(' ')
    l=[]
    if ii>0:
        if prob[ii-1][ji]!=' ':
            l.append('d')
    if ii<2:
        if prob[ii+1][ji]!=' ':
            l.append('u')
    if ji>0:
        if prob[ii][ji-1]:
            l.append('r')
    if ji<2:
        if prob[ii][ji+1]:
            l.append('l')
    
    for i in l:
        prob=convertlist(prob)
        prob2=copy.deepcopy(prob)
        if i=='d':
            t=prob2[ii][ji]
            prob2[ii][ji]=prob2[ii-1][ji]
            prob2[ii-1][ji]=t
        if i=='u':
            t=prob2[ii][ji]
            prob2[ii][ji]=prob2[ii+1][ji]
            prob2[ii+1][ji]=t
        if i=='r':
            t=prob2[ii][ji]
            prob2[ii][ji]=prob2[ii][ji-1]
            prob2[ii][ji-1]=t
        if i=='l':
            t=prob2[ii][ji]
            prob2[ii][ji]=prob2[ii][ji+1]
            prob2[ii][ji+1]=t
        prob=converttup(prob)
        prob2=converttup(prob2)
        if prob2 not in graph:
            graph[prob].append(prob2)
            nodes.append(prob2)
            graph[prob2]=[]
            f[prob2]=checkh(prob2)+depth
        if prob2==soln:
            bfs()
    prob3=prob2
    min=10000000
    for prob2 in graph[prob]:
        if f[prob2]<=min:
            min=f[prob2]
            prob3=prob2
    solve(prob3)        
    
print("Enter problem: ")
generate()
print("Problem is: ")
display(prob1)
print("Enter soln: ")
generatesoln()
print("Soln is: ")
display(soln)
solve(prob1)
```
