```python
import re
import gensim.models
from gensim.models.callbacks import CallbackAny2Vec
from gensim.models import KeyedVectors
```

Loading Sentence Piece encoded sentences

```python
sentencessp=[]
with open('malayalam_senteces_sp2.txt','r',encoding='utf-8') as file:
            for line in file:
                line = re.sub(r'[^\u0D00-\u0D7F\s]', ' ', line)
                sentencessp += [[w for w in line.split()]]
```

Loading og sentences

```python
sentencesog=[]
with open('malayalam_senteces2.txt','r',encoding='utf-8') as file:
            for line in file:
                line = re.sub(r'[^\u0D00-\u0D7F\s]', ' ', line)
                sentencesog += [[w for w in line.split()]]
```

Loading the original words

```python
og_words=[]
for line in sentencesog:
    #print(line)
    for word in line:
        #print(word)
        og_words.append(word)
og_words=list(set(og_words))
```

Optional Callback for gensim

```python
class callable(CallbackAny2Vec):
    def __init__(self) -> None:
        super().__init__()
        self.words=[re.sub(r'[^\u0D00-\u0D7F\s]', '', 'എടുക്കാത്തതാകുന്നു'),re.sub(r'[^\u0D00-\u0D7F\s]', '', 'ദൈവം'),re.sub(r'[^\u0D00-\u0D7F\s]', '', 'പുണ്യസ്ഥലം')
                    ,re.sub(r'[^\u0D00-\u0D7F\s]', '', 'ചീത്ത'),re.sub(r'[^\u0D00-\u0D7F\s]', '', 'അച്ഛന്‍'),re.sub(r'[^\u0D00-\u0D7F\s]', '', 'ദുഷ്ടന്‍'),
                    re.sub(r'[^\u0D00-\u0D7F\s]', '', 'പദാര്ത്ഥം'),re.sub(r'[^\u0D00-\u0D7F\s]', '', 'ചുവന്ന')]
    def on_train_end(self, model):
        for word in self.words:
            print(word,": ",[x[0] for x in model.wv.most_similar(word)])

callback=callable()
```

Training gensim

```python
model1= gensim.models.Word2Vec(sentences=sentencesog,callbacks=[callback],min_count=1,vector_size=300,window=25,sg=1,negative=15,ns_exponent =1,seed=10,workers=16,epochs=5)
```

```python
model2= gensim.models.Word2Vec(sentences=sentencesog,min_count=1,vector_size=300,window=60,sg=1,negative=30,ns_exponent =1,seed=10,workers=16,epochs=15)
```

```python
import numpy as np
og_vectors={}
for word in og_words:
    vector=np.zeros((1,300))
    encodeds=sp.EncodeAsPieces(word)#re.sub(r'[^\u0D00-\u0D7F\s]', '', word))
    for encoded in encodeds:
        try:
            encoded=re.sub(r'[^\u0D00-\u0D7F\s]', '', encoded)
            if encoded:
                vector+=model2[encoded]
        except:
            print(word,encoded)
    og_vectors[word]=vector
```

```python
import sentencepiece as spm
from gensim import similarities
from sklearn.metrics.pairwise import cosine_similarity
sp = spm.SentencePieceProcessor(model_file='malayalam_sent/mala_sp_bpe.model')
wordog="സഹോദരി"
word_vector=og_vectors[re.sub(r'[^\u0D00-\u0D7F\s]', '', wordog)]
#print(word_vector)
for word in list(og_words):
    if wordog!=word:
        similarity = cosine_similarity(word_vector, og_vectors[word])
        if similarity>0.70:
            print(f"Similarity between {wordog} and {word}: {similarity[0][0]}")
```

```python
import sentencepiece as spm
sp = spm.SentencePieceProcessor(model_file='malayalam_sent/mala_sp_bpe.model')
word="അന്യായം"
encodeds = sp.EncodeAsPieces(word)
similar2=model2.most_similar(encodeds,topn=25)
similar2=[x[0] for x in similar2]
sp.DecodePieces(similar2)
print(similar2)
similar2=[re.sub(r'[^\u0D00-\u0D7F\s]', '', x) for x in similar2]
print(similar2)
#escaped_words = [re.escape(word) for word in og_words]
#print(escaped_words)
match=[]
for word in similar2:
    for x in og_words:
        if word in x:
            match.append(x)
print(match)
```
